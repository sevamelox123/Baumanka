```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def activation_func_hidden(x):
    return 0 if x < 0 else 1

def activation_func_final(x):
    return 0 if x < 0 else 1
```

```python
class Neuron:
    def __init__(self, weights: np.array, act_func) -> None:
        self.weights = weights
        self.act_func = act_func

    def calc(self, x: np.array):
        summ = np.dot(self.weights, x)
        y = self.act_func(summ)
        return y
```

```python
N = 100
x1 = np.random.random(N)
x2 = np.random.random(N)
x3 = np.array([1 for _ in range(N)])
X = np.array([x1, x2, x3])
```

```python
hidden_layer = []
final_layer = []

w1 = np.array([1, 1, -1.5])
w2 = np.array([1, 1, -0.5])
wf = np.array([1, -1])

hidden_layer.append(Neuron(w1, activation_func_hidden))
hidden_layer.append(Neuron(w2, activation_func_hidden))
final_layer.append(Neuron(wf, activation_func_final))

res = []
for i in range(N):
    _res = []
    for j in range(len(final_layer)):
        _res.append(final_layer[j].calc([ _.calc(np.array([x1[i], x2[i], x3[i]])) for _ in hidden_layer]))
    res.append(_res)
```

```python
l1x = [0, 0.5]
l1y = [0.5, 0]
l2x = [0.5, 1]
l2y = [1, 0.5]

fig = plt.figure(figsize=(5, 5))
plt.axis([0, 1, 0, 1])

for i in range(len(res)):
    if res[i] == [1]:
        plt.scatter(x1[i], x2[i], c = 'blue')
    else:
        plt.scatter(x1[i], x2[i], c = 'red')

plt.plot(l1x, l1y, c = 'black')
plt.plot(l2x, l2y, c = 'black')
plt.show()
```
